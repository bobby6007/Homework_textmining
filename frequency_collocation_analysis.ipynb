```python
import ujson

CHK_POSES = {"NNG", "NNP", "VV", "VA" "XR"} # 명사와 어근
INPUT_FILE_NAME = "/content/mayor_news_morph.jsonl"

publishers = []
mayors = []
documents = []

mayor_idxes = {}


with open(INPUT_FILE_NAME, 'r', encoding='utf-8') as input_file:
    for idx, line in enumerate(input_file):
        data = ujson.loads(line)


        publishers.append(data.get('publisher', None))
        bodies.append(data.get('body', None))
        mayors.append(data.get('mayor', None))
        mayor_idxes.update({data["mayor"]: idx})


        word_mas = []
        body_mas = data["body_ma"]
        # 문장별로 가져온다.
        for body_ma in body_mas:
            word_ma = [lex + "다" if pos in {"VA", "VV"} else lex for lex, pos in body_ma if pos in CHK_POSES]
            word_mas.append(word_ma)

        documents.append(word_mas)

articles = {
    "publisher": publishers,
    "mayor": mayors,
    "wordset":documents
}

```
